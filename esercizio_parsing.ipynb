```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.greenme.it/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
bs = BeautifulSoup(response.text, "lxml")

dict_news = {"link": [], "titolo": []}

articoli = bs.find_all('article')
print(len(articoli))

for art in articoli:
    link_tag = art.find('a', href=True)
    titolo_tag = art.find(['h2', 'h3'])

    if link_tag and titolo_tag:
        dict_news["link"].append(link_tag['href'])
        dict_news["titolo"].append(titolo_tag.get_text(strip=True))

df_news = pd.DataFrame(dict_news)
print(f"Articoli trovati: {len(df_news)}")
df_news.head()
```

```python
dict_news = {"link": [], "titolo": [], "autore": [], "data": []}
url = 'https://www.greenme.it/'
response = requests.get(url)
bs = BeautifulSoup(response.text,"lxml")

temp = bs.find_all('a', 'image-wrapper')
for i in temp:
dict_news["link"].append(i.get('href'))
dict_news["titolo"].append(i.get('title'))

temp = bs.find_all('div', 'post-meta')
for i in temp:
dict_news["autore"].append(i.find('span','post-author'))
dict_news["data"].append(i.find('span','post-date'))


df_news = pd.DataFrame(dict_news, columns=["link", "titolo", "autore", "data"])

df_news
```
